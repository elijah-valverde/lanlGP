# Particulate Matter at M1
Here we overlay the PM10 concentration with environmental conditions {humidity, temperature, & pressure}


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
pm_data_m1 = pd.read_csv(r'C:\Users\396760\lanl_data\data\PM_clean\ARM0810.SAIL_PM_M1.20240509_14_05_423_cleaned.csv', delimiter='\t')

# Convert timestamp to datetime
pm_data_m1['Time(UTC)'] = pd.to_datetime(pm_data_m1['Time(UTC)'])

# Set the timestamp as index
pm_data_m1.set_index('Time(UTC)', inplace=True)

# Plot Relative Humidity with PM10 concentration
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('PM10 Concentration (µg/m³)', color=color)
ax1.plot(pm_data_m1.index, pm_data_m1['pm_10_ug_per_m3'], color=color, label='PM10')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:green'
ax2.set_ylabel('Relative Humidity (%)', color=color)  # we already handled the x-label with ax1
ax2.plot(pm_data_m1.index, pm_data_m1['sample_rh_pct'], color=color, label='Relative Humidity', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('PM10 Concentration and Relative Humidity Over Time (M1)')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.grid(True)
plt.show()

```

```python
# Plot Temperature with PM10 concentration
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('PM10 Concentration (µg/m³)', color=color)
ax1.plot(pm_data_m1.index, pm_data_m1['pm_10_ug_per_m3'], color=color, label='PM10')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel('Temperature (°C)', color=color)  # we already handled the x-label with ax1
ax2.plot(pm_data_m1.index, pm_data_m1['sample_temp_C'], color=color, label='Temperature', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('PM10 Concentration and Temperature Over Time (M1)')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.grid(True)
plt.show()
```

```python
# Plot Pressure with PM10 concentration
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('PM10 Concentration (µg/m³)', color=color)
ax1.plot(pm_data_m1.index, pm_data_m1['pm_10_ug_per_m3'], color=color, label='PM10')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:purple'
ax2.set_ylabel('Pressure (mmHg)', color=color)  # we already handled the x-label with ax1
ax2.plot(pm_data_m1.index, pm_data_m1['sample_pres_mmHg'], color=color, label='Pressure', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

# Set secondary y-axis range and ticks
ax2.set_ylim(0, 1000)
ax2.set_yticks(range(0, 1001, 125))

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('PM10 Concentration and Pressure Over Time (M1)')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.grid(True)
plt.show()
```
