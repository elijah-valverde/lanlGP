# Particulate Matter at M1
Here we overlay the PM1 concentration with environmental conditions {humidity, temperature, & pressure}


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
pm_data_m1 = pd.read_csv(r"C:\Users\396760\lanl_data\PM_clean\ARMSAILM1_cleaned.csv")

# Convert timestamp to datetime
pm_data_m1['Time(UTC)'] = pd.to_datetime(pm_data_m1['Time(UTC)'])

# Plot Relative Humidity with PM1 concentration
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:blue'
ax1.set_xlabel('Minute Frequency')
ax1.set_ylabel('PM1 Concentration (µg/m³)', color=color)
ax1.plot(pm_data_m1['Time(UTC)'], pm_data_m1['pm_1_ug_per_m3'], color=color, label='PM1')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:green'
ax2.set_ylabel('Relative Humidity (%)', color=color)  # we already handled the x-label with ax1
ax2.plot(pm_data_m1['Time(UTC)'], pm_data_m1['sample_rh_pct'], color=color, label='Relative Humidity', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('PM1 Concentration and Relative Humidity Over Time (M1)')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.grid(True)
plt.show()
```

```python
# Plot Temperature with PM1 concentration
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:blue'
ax1.set_xlabel('Minute Frequency')
ax1.set_ylabel('PM1 Concentration (µg/m³)', color=color)
ax1.plot(pm_data_m1['Time(UTC)'], pm_data_m1['pm_1_ug_per_m3'], color=color, label='PM1')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel('Temperature (°C)', color=color)  # we already handled the x-label with ax1
ax2.plot(pm_data_m1['Time(UTC)'], pm_data_m1['sample_temp_C'], color=color, label='Temperature', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('PM1 Concentration and Temperature Over Time (M1)')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.grid(True)
plt.show()
```

```python
# Plot Pressure with PM1 concentration
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:blue'
ax1.set_xlabel('Minute Frequency')
ax1.set_ylabel('PM1 Concentration (µg/m³)', color=color)
ax1.plot(pm_data_m1['Time(UTC)'], pm_data_m1['pm_1_ug_per_m3'], color=color, label='PM1')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:purple'
ax2.set_ylabel('Pressure (mmHg)', color=color)  # we already handled the x-label with ax1
ax2.plot(pm_data_m1['Time(UTC)'], pm_data_m1['sample_pres_mmHg'], color=color, label='Pressure', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

# Set secondary y-axis range and ticks
ax2.set_ylim(0, 1000)
ax2.set_yticks(range(0, 1001, 125))

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('PM1 Concentration and Pressure Over Time (M1)')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.grid(True)
plt.show()
```

```python
# Load the data
pm_data_m1 = pd.read_csv(r"C:\Users\396760\lanl_data\PM_clean\ARMSAILM1_cleaned.csv")

# Adjust display settings
pd.set_option('display.max_rows', None)   # Show all rows
pd.set_option('display.max_columns', None) # Show all columns
pd.set_option('display.max_colwidth', None) # Show full column width
pd.set_option('display.width', None)      # Auto-detect the display width

# Convert timestamp to datetime
pm_data_m1['Time(UTC)'] = pd.to_datetime(pm_data_m1['Time(UTC)'])

# Set the timestamp as index
pm_data_m1.set_index('Time(UTC)', inplace=True)

# Function to get top 30 maximum observations
def get_top_30_max(df, column_name):
    top_30 = df.nlargest(50, column_name)[[column_name]]
    return top_30

# Get top 30 observations for PM1, PM2.5, and PM10
top_30_pm1 = get_top_30_max(pm_data_m1, 'pm_1_ug_per_m3')
top_30_pm25 = get_top_30_max(pm_data_m1, 'pm_25_ug_per_m3')
top_30_pm10 = get_top_30_max(pm_data_m1, 'pm_10_ug_per_m3')

# Display results
print("Top 100 PM1 Observations:")
print(top_30_pm1)

print("\nTop 100 PM2.5 Observations:")
print(top_30_pm25)

print("\nTop 100 PM10 Observations:")
print(top_30_pm10)
```
