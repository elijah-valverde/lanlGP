# Particulate Matter at S2
Here we overlay the PM2.5 concentration with environmental conditions {humidity, temperature} with different averages {minute, hourly, daily}. We omit the pressure as the data given had pressure at a constant value of 0.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
pm_data_s2 = pd.read_csv(r'C:\Users\396760\lanl_data\PM_clean\ARMSAILS2_cleaned.csv')

# Convert timestamp to datetime
pm_data_s2['Time(UTC)'] = pd.to_datetime(pm_data_s2['Time(UTC)'])

# Set the timestamp as index
pm_data_s2.set_index('Time(UTC)', inplace=True)

# Plot Relative Humidity with PM2.5 concentration
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('PM2.5 Concentration (µg/m³)', color=color)
ax1.plot(pm_data_s2.index, pm_data_s2['pm_25_ug_per_m3'], color=color, label='PM2.5')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:green'
ax2.set_ylabel('Relative Humidity (%)', color=color)  # we already handled the x-label with ax1
ax2.plot(pm_data_s2.index, pm_data_s2['sample_rh_pct'], color=color, label='Relative Humidity', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('PM2.5 Concentration and Relative Humidity Over Time (S2)')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.grid(True)
plt.show()
```

```python
# Plot Temperature with PM2.5 concentration
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('PM2.5 Concentration (µg/m³)', color=color)
ax1.plot(pm_data_s2.index, pm_data_s2['pm_25_ug_per_m3'], color=color, label='PM2.5')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel('Temperature (°C)', color=color)  # we already handled the x-label with ax1
ax2.plot(pm_data_s2.index, pm_data_s2['sample_temp_C'], color=color, label='Temperature', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('PM2.5 Concentration and Temperature Over Time (S2)')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.grid(True)
plt.show()
```
